```python

def marker_postprocess(marker, threshold=0.9):
    marker0 = opening(marker,disk(15))
#     marker = closing(marker,disk(5))
    marker1 = (marker0 > threshold)
    marker = (marker>0.5)
    marker_temp = label(marker)
    marker2 = erosion(marker1, disk(9))
#     temp_mask = 
    marker3 = remove_small_holes(marker2)
#     markers = label(marker4)
#     marker4 = opening(marker3, disk(5))
#     marker5 = closing(marker4, disk(5))
    for i in range(1,np.max(marker_temp)+1):
        temp = (marker_temp == i).astype(np.uint8)
        if not (temp*marker3).any():
            marker3 = marker3 + temp
    markers = label(marker3)
#     print(np.max(markers))
    return markers


def border_postprocess(border, threshold=0.15):
    border1 = (border > threshold)
    border1 = border1.astype(np.uint8)
    border2 = closing(border1)
    return border2


def mini_postprocess(final_mask, origin_mask):
    # 添加那些被移除的小细胞
    # final_mask1 = erosion(final_mask, disk(5))
    # final_mask2 = opening(final_mask1, disk(3))
    water_mask = (final_mask > 0).astype(np.uint8)
    dropped = origin_mask - water_mask
    dropped = label(dropped)
    dropped = np.where(dropped, dropped + final_mask.max(), 0)
#     final_mask1 = final_mask + dropped
    # 去除那些很小的杂点
    remove_mask = remove_small_objects(final_mask, 100)
    final_mask = final_mask + dropped
    out = np.where(remove_mask, final_mask, 0)
#     if np.max(out) > 30:
#         out = erosion(out, disk(3))
#     out = erosion(out, disk(3))
    region = regionprops(out)
    num = len(region)
    max_area = 0
    for i in range(num):
        max_area = max(max_area, region[i].area)
    for i in range(num):
        row,col = region[i].centroid
#         print(row,col)
        curr_label = out[int(row),int(col)]
        if region[i].area < 0.4*max_area:
#             temp = (region[i].area).astype(np.uint8)
            temp = (out == curr_label).astype(np.uint8)
            temp = dilation(temp, disk(9))
            out = np.where(temp, curr_label,out)
        elif region[i].area < 0.6*max_area:
            temp = (out == curr_label).astype(np.uint8)
            temp = dilation(temp, disk(5))
            out = np.where(temp, curr_label,out)            
    if np.max(out) > 23:
        out = erosion(out, disk(3))
        out = closing(out, disk(3))
#     else:
#         out = dilation(out, disk(3))
#     elif np.max(out) < 15:
#         out = dilation(out, disk(5))
#     # out = opening(out, disk(5))
    return out


def output_postprocesss(marker, border):
    # plt.figure()
    # plt.imshow(border)
    marker_new = marker_postprocess(marker)
    border_new = border_postprocess(border)
    
#     plt.figure()
#     plt.imshow(border_new)
#     plt.figure()
#     plt.imshow(marker_new)
    mask_all = (marker > 0.5).astype(np.uint8)
#     border = find_boundaries(mask_all)
    # border_new = dilation(border_new, disk(3))
    mask_all = np.maximum(mask_all, border_new)
#     plt.figure()
#     plt.imshow(mask_all)
    mask_all = remove_small_holes(mask_all, 100)
#     mask_all = dilation(mask_all, disk(7))
#     plt.figure()
#     plt.imshow(mask_all)
    # 让有label的地方是最小值
#     water_all = np.maximum(255 - border_new, 255 - marker_new)
    water_all = 255 - marker_new
#     water_all = ndi.distance_transform_edt(mask_all)
    out = watershed(water_all, marker_new, mask=mask_all)
    # plt.figure()
    # plt.imshow(out)
    out = mini_postprocess(out, mask_all)
    return out
```

```python
def cal_jaccard(real, pred):

#     real = (255*real).astype(np.uint8)
    label_num = np.max(real)
#     print(label_num)
    score = 0
    num = 0
    for i in range(label_num + 1):
        pos_i = (real == i)
        a = np.bincount(pred[pos_i])
        pred_label = 1000000
#         print(a.shape)
        if a.shape[0] > 0: pred_label = np.argmax(a)
        pred_tmp = np.where(pred == pred_label, 1, 0)
        real_tmp = np.where(real == i, 1, 0)
        if real_tmp.any():
            num += 1
            js = np.count_nonzero((pred_tmp * real_tmp) == 1)/np.count_nonzero((pred_tmp + real_tmp) >= 1)
            score += js
    return score/num
```
